```python
import json
import random

# ğŸ”§ ê²½ë¡œ ì„¤ì •
train_json_path = r"C:\Users\Lenovo\Study\2025_Samsung_AI_Challenge\dataset\VQAv2\train.json"
val_json_path = r"C:\Users\Lenovo\Study\2025_Samsung_AI_Challenge\dataset\VQAv2\val.json"
sampled_train_json_path = r"C:\Users\Lenovo\Study\2025_Samsung_AI_Challenge\dataset\VQAv2\train_10000_seed42.json"
sampled_json_path = r"C:\Users\Lenovo\Study\2025_Samsung_AI_Challenge\dataset\VQAv2\val_2000_seed42.json"

# âœ… ì‹œë“œ ê³ ì •
random.seed(42)

# ğŸ“‚ ì›ë³¸ JSON ë¡œë“œ
with open(train_json_path, "r", encoding="utf-8") as f:
    train_data = json.load(f)

with open(val_json_path, "r", encoding="utf-8") as f:
    val_data = json.load(f)
    
# ğŸ” Train 10000ê°œ, Validation 2000ê°œ ìƒ˜í”Œë§
sampled_train_data = random.sample(train_data, 10000)
sampled_val_data = random.sample(val_data, 2000)

# ğŸ’¾ ìƒ˜í”Œ ì €ì¥
with open(sampled_train_json_path, "w", encoding="utf-8") as f:
    json.dump(sampled_train_data, f, indent=2, ensure_ascii=False)

with open(sampled_json_path, "w", encoding="utf-8") as f:
    json.dump(sampled_val_data, f, indent=2, ensure_ascii=False)

print(f"âœ… Train 10000ê°œ ìƒ˜í”Œ ì €ì¥ ì™„ë£Œ: {sampled_train_json_path}")
print(f"âœ… Validation 2000ê°œ ìƒ˜í”Œ ì €ì¥ ì™„ë£Œ: {sampled_json_path}")
```
